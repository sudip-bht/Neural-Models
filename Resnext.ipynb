```python
import torch
import torchvision
from torchvision import datasets, transforms, models
import torch.nn as nn
from torch.nn.modules.loss import BCEWithLogitsLoss
from torch.optim import lr_scheduler
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
```

```python

def main():
    # Directories
    traindir = "/kaggle/input/ham10000-augmented-dataset/Data folder/train"
    testdir = "/kaggle/input/ham10000-augmented-dataset/Data folder/train"

    train_transforms = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        torchvision.transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225],
        ),
    ])
    test_transforms = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        torchvision.transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225],
        ),
    ])

    train_data = datasets.ImageFolder(traindir, transform=train_transforms)
    test_data = datasets.ImageFolder(testdir, transform=test_transforms)
    trainloader = torch.utils.data.DataLoader(train_data, shuffle=True, batch_size=16, num_workers=4, pin_memory=True)
    testloader = torch.utils.data.DataLoader(test_data, shuffle=False, batch_size=16, num_workers=4, pin_memory=True)

    # Training step function
    def make_train_step(model, optimizer, loss_fn):
        def train_step(x, y):
            model.train()
            yhat = model(x)
            loss = loss_fn(yhat, y)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            return loss
        return train_step

    # Device
    device = "cuda" if torch.cuda.is_available() else "cpu"

    # Model setup
    model = models.resnext50_32x4d(pretrained=True)
    for param in model.parameters():
        param.requires_grad = False
    nr_filters = model.fc.in_features
    model.fc = nn.Linear(nr_filters, 1)
    model = model.to(device)

    # Loss and optimizer
    loss_fn = BCEWithLogitsLoss()
    optimizer = torch.optim.Adam(model.fc.parameters())

    # Train step
    train_step = make_train_step(model, optimizer, loss_fn)

    # Training and validation loop
    n_epochs = 50
    early_stopping_tolerance = 3
    early_stopping_threshold = 0.03
    early_stopping_counter = 0
    best_loss = float('inf')
    best_model_wts = model.state_dict()

    epoch_train_losses = []
    epoch_test_losses = []
    epoch_accuracies = []
    epoch_precisions = []
    epoch_recalls = []
    epoch_f1_scores = []
    epoch_sensitivities = []
    epoch_specificities = []

    for epoch in range(n_epochs):
        # Training phase
        model.train()
        epoch_loss = 0.0
        for x_batch, y_batch in tqdm(trainloader, total=len(trainloader)):
            x_batch = x_batch.to(device)
            y_batch = y_batch.unsqueeze(1).float().to(device)
            loss = train_step(x_batch, y_batch)
            epoch_loss += loss.item()

        epoch_train_losses.append(epoch_loss / len(trainloader))

        # Validation phase
        model.eval()
        all_preds = []
        all_labels = []
        with torch.no_grad():
            epoch_test_loss = 0.0
            for x_batch, y_batch in testloader:
                x_batch = x_batch.to(device)
                y_batch = y_batch.unsqueeze(1).float().to(device)
                yhat = model(x_batch)
                loss = loss_fn(yhat, y_batch)
                epoch_test_loss += loss.item()

                preds = torch.sigmoid(yhat).cpu().numpy() >= 0.5
                labels = y_batch.cpu().numpy()
                all_preds.extend(preds)
                all_labels.extend(labels)

        epoch_test_losses.append(epoch_test_loss / len(testloader))

        # Calculate evaluation metrics
        accuracy = accuracy_score(all_labels, all_preds)
        precision = precision_score(all_labels, all_preds)
        recall = recall_score(all_labels, all_preds)  # Sensitivity
        f1 = f1_score(all_labels, all_preds)

        # Confusion matrix
        cm = confusion_matrix(all_labels, all_preds)
        tn, fp, fn, tp = cm.ravel()
        sensitivity = tp / (tp + fn)  # Sensitivity
        specificity = tn / (tn + fp)  # Specificity

        epoch_accuracies.append(accuracy)
        epoch_precisions.append(precision)
        epoch_recalls.append(recall)
        epoch_f1_scores.append(f1)
        epoch_sensitivities.append(sensitivity)
        epoch_specificities.append(specificity)

        # Print epoch results
        print(f'Epoch {epoch+1}/{n_epochs}')
        print(f'Train Loss: {epoch_train_losses[-1]:.4f}')
        print(f'Val Loss: {epoch_test_losses[-1]:.4f}')
        print(f'Accuracy: {accuracy:.4f}')
        print(f'Precision: {precision:.4f}')
        print(f'Recall (Sensitivity): {recall:.4f}')
        print(f'F1 Score: {f1:.4f}')
        print(f'Specificity: {specificity:.4f}\n')

        # Early stopping
        # if epoch_test_losses[-1] < best_loss:
        #     best_loss = epoch_test_losses[-1]
        #     best_model_wts = model.state_dict()
        #     early_stopping_counter = 0
        # else:
        #     early_stopping_counter += 1

        # if early_stopping_counter >= early_stopping_tolerance or best_loss <= early_stopping_threshold:
        #     print("\nTerminating: early stopping")
        #     break

    # Load best model
    model.load_state_dict(best_model_wts)

    # Plotting
    plt.figure(figsize=(15, 10))

    plt.subplot(2, 2, 1)
    plt.plot(range(1, len(epoch_train_losses) + 1), epoch_train_losses, label='Training Loss')
    plt.plot(range(1, len(epoch_test_losses) + 1), epoch_test_losses, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(2, 2, 2)
    plt.plot(range(1, len(epoch_accuracies) + 1), epoch_accuracies, label='Accuracy')
    plt.plot(range(1, len(epoch_precisions) + 1), epoch_precisions, label='Precision')
    plt.plot(range(1, len(epoch_recalls) + 1), epoch_recalls, label='Recall')
    plt.plot(range(1, len(epoch_f1_scores) + 1), epoch_f1_scores, label='F1 Score')
    plt.xlabel('Epochs')
    plt.ylabel('Metrics')
    plt.legend()

    plt.subplot(2, 2, 3)
    plt.plot(range(1, len(epoch_sensitivities) + 1), epoch_sensitivities, label='Sensitivity')
    plt.plot(range(1, len(epoch_specificities) + 1), epoch_specificities, label='Specificity')
    plt.xlabel('Epochs')
    plt.ylabel('Metrics')
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Benign", "Malignant"])
    disp.plot()
    plt.show()

    # Create a DataFrame for the results
    results_df = pd.DataFrame({
        'Epoch': range(1, len(epoch_train_losses) + 1),
        'Train Loss': epoch_train_losses,
        'Val Loss': epoch_test_losses,
        'Accuracy': epoch_accuracies,
        'Precision': epoch_precisions,
        'Recall (Sensitivity)': epoch_recalls,
        'F1 Score': epoch_f1_scores,
        'Sensitivity': epoch_sensitivities,
        'Specificity': epoch_specificities
    })

    # Print the DataFrame
    print(results_df)

    # Optionally save the DataFrame to a CSV file
    results_df.to_csv('training_results.csv', index=False)

    # Inference function
    def inference(test_data, model, device):
        model.eval()
        idx = torch.randint(0, len(test_data), (1,)).item()
        sample = torch.unsqueeze(test_data[idx][0], dim=0).to(device)
        with torch.no_grad():
            output = torch.sigmoid(model(sample))
        print("Prediction:", "Malignant" if output >= 0.5 else "Benign")
        plt.imshow(test_data[idx][0].permute(1, 2, 0))
        plt.show()

    inference(test_data, model, device)
    torch.save(model, '/kaggle/working/ResneXt')

if __name__ == '__main__':
    main()
```
